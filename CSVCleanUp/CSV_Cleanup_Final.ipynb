```python
import pandas as pd
```

## Highway Vehicle Miles Traveled

```python
# File to Load (Remember to Change These)
file_to_load = "../ETL_CSV_Files/Highway_Vehicle_Miles_Traveled.csv"

# Read File and store into Pandas data frame
miles_traveled = pd.read_csv(file_to_load)
miles_traveled.head()
```

```python
# Keep rows were there is at least 3 values that are not NaN.
# This makes it so that Index and Date are kept, but at least one other column needs to have a value other than NaN to 
# stay in the dataframe
columnNames=miles_traveled.columns
df=miles_traveled[columnNames].dropna(thresh=3)
df.head()
```

```python
# Rename columns for ease
df.rename(columns = {'Index':'IndexID',
                         'Highway Vehicle Miles Traveled - All Systems':'HVMT_All_Systems', 'Highway Vehicle Miles Traveled - Total Rural':'HVMT_Total_Rural', 
                              'Highway Vehicle Miles Traveled - Other Rural':'HVMT_Other_Rural',
                                'Highway Vehicle Miles Traveled - Rural Other Arterial':'HVMT_Rural_Other_Arterial',
                                'Highway Vehicle Miles Traveled - Rural Interstate':'HVMT_Rural_Interstate'}, inplace = True)
```

```python
df.head()
```

```python
# save to new CSV
df_csv_data = df.to_csv('Cleaned_Highway_Vehicle_Miles_Traveled.csv', index = True)
```

## US Air

Us Air gets split into Airline Traffic and Carrier Cargo

```python
# File to Load (Remember to Change These)
file_to_load = "../ETL_CSV_Files/US_Air.csv"

# Read File and store into Pandas data frame
US_Air = pd.read_csv(file_to_load)
US_Air.head()
```

## Airline Traffic

```python
airline=US_Air[["Index","Date","U.S. Airline Traffic - Total - Seasonally Adjusted","U.S. Airline Traffic - International - Seasonally Adjusted","U.S. Airline Traffic - Domestic - Seasonally Adjusted"]]
airline.head()
```

```python
# Keep rows were there is at least 3 values that are not NaN.
# This makes it so that Index and Date are kept, but at least one other column needs to have a value other than NaN to 
# stay in the dataframe
columnNames=airline.columns
airline_df=airline[columnNames].dropna(thresh=3)
airline_df.head()
```

```python
# save to new CSV
airline_csv = airline_df.to_csv('Cleaned_Airline_Traffic.csv', index = True)
```
